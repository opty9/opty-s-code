```python
import seaborn as sb
```

```python
import pandas as pd 
```

```python
# drink.csv 파일 로드 

drinks = pd.read_csv('../csv/drinks.csv')
drinks.head(5)
```

```python
# 상관계수 행렬 출력 

corr = drinks [
    [
     'beer_servings',
     'spirit_servings',
     'wine_servings',
     'total_litres_of_pure_alcohol'
    ]
    ].corr(method='pearson')
corr
```

```python
# heatmap 출력 


cols_view = ['beer' ,'spirit', 'wine', 'alcohol']

sb.set(font_scale=1.5)

hm = sb.heatmap(
    corr.values,
    cbar = True, # 색상에 대한 설명
    annot = True, # 네모칸 안 수치 유무     
    square = True, # 별 의미없는듯 / 직사각형 정사각형 여부
    fmt = '.2f', # 소수점 몇자리까지 표시인지
    annot_kws = {'size' : 20}, # 글자 크기 
    xticklabels = cols_view, 
    yticklabels = cols_view
)
plt.tight_layout()
plt.show()
```

```python
sb.set(style='whitegrid', context = 'notebook')

sb.pairplot(
    drinks[
    ['beer_servings',
     'spirit_servings',
     'wine_servings',
     'total_litres_of_pure_alcohol'
    ]
    ],
    height = 5
)
plt.show()
```
