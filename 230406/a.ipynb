```python
import pandas as pd 
```

```python
### dataframe의 결합

#단순결합 - 단순 행추가, 열추가 / contact() 이라는 함수를 사용하여 데이터프레임 결합 / 매개변수 axis : 행을 결합, 열을 결합
#매개변수 ignore_index : 데이터프레임을 결합한 뒤 인덱스를 초기화 할 것인지 지정

#특정 조건에 맞게 결합 - 데이터프레임에 특정한 조건이 참이면 데이터를 결합하는 형태 / 열을 추가하는 함수로 merge()
#매개변수 on : 특정한 조건 지정 
#매개변수 how : (left, right, inner, full) 데이터프레임의 기준을 지정 
```

```python
# 데이터프레임 생성

df = pd.DataFrame({
    'S1' : [1,2,3,4],
    'S2' : [5,6,7,8]
})
df
```

```python
df2 = pd.DataFrame([1,2])
df2
```

```python
# concat 함수를 이용해 데이터프레임 결합
# 단순하게 행 추가 

                            #"rows"
#pd.concat([df, df2], axis = 0, ignore_index = True)

pd.concat([df, df2], axis = 0).reset_index(drop=True)
```

```python
df3 = pd.DataFrame({
    'S1' : [1,2]
})
df3
```

```python
pd.concat([df, df3], axis = 0)
```

```python
#단순하게 열을 결합
#axis = 0->1

pd.concat([df, df2], axis = 1)
```

```python
df2.index = [6,7]

df2
```

```python
pd.concat([df, df3], axis = 1)
```

```python
pd.concat([df,df2,df3], axis= 0)
```

```python
#merge()
#특정한 조건에 맞춰서 열을 추가하는 함수 

df = pd.DataFrame({
    "id" : [1,2,3,4,5],
    'item' : ['a', 'b', 'a', 'a', 'b']
})

df2 = pd.DataFrame ({
    'id' : [1,4,5,2,3], 
    'name' : ['kim', 'lee', 'park', 'choi', 'ha']
})

df2
```

```python
pd.concat([df,df2], axis=0)
```

```python
#merge 함수로 차이알기

pd.merge(df, df2, on = 'id', how = 'left')
```

```python
df3 = pd.DataFrame({
    'item' : [100,200,300]
}) 
df3


    
```

```python
pd.merge(df, df3, on ='item', how='right')

```

```python
df.loc[3, 'item'] = 'd'
df
```

```python
pd.merge(df, df3, on='item', how = 'outer')
```
